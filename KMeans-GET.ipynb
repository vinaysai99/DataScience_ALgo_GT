```python
# KMeans simulation in python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#create some data

from sklearn.datasets import make_blobs
```

```python
help(make_blobs)
```

```python
#visualize data

data = make_blobs(n_samples=200, centers=4, n_features=2, cluster_std= 1.8, random_state=0)
```

```python
#Important KMeans

from sklearn.cluster import KMeans
```

```python
# Choose number of Clusters
kmeans = KMeans(n_clusters = 4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
f, (ax1,ax2) = plt.subplots(1, 2, sharey = True, figsize =(10,6))
ax1.set_title("K Means")
ax1.scatter(data[0][:,0],data[0][:,1], c=kmeans.labels_, cmap = 'rainbow')
ax2.set_title("Original Data")
ax2.scatter(data[0][:,0],data[0][:,1], c=data[1], cmap = 'rainbow')
ax1.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 100, c = 'black', label = 'Centroids')
```

```python
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    #kmeanModel.fit(df)
    kmeanModel.fit(data[0])
    distortions.append(kmeanModel.inertia_)
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```
